## Clusterização

### Inicialização da matriz de pertinência

```python
def inicializacao_matriz_pertinencia(num_amostras, num_clusters):
    matriz_pertinencia = np.random.rand(num_amostras, num_clusters)  # Gera matriz aleatória inicial
    matriz_pertinencia = matriz_pertinencia / matriz_pertinencia.sum(axis=1, keepdims=True)  # Normaliza para garantir que a soma dos graus de pertinência seja 1
    return matriz_pertinencia
```

### Atualização dos medoids (Corrigir)

```python
def atualizacao_medoids(dados, matriz_pertinencia, m):
    num_clusters = matriz_pertinencia.shape[1]
    medoids = np.zeros((num_clusters, dados.shape[1]))

    for i in range(num_clusters):
        ponderacao = matriz_pertinencia[:, i] ** m
        distancias_total = np.sum(np.linalg.norm(dados[:, np.newaxis] - dados, axis=2) * ponderacao[:, np.newaxis], axis=0)
        medoids[i] = dados[np.argmin(distancias_total)]

    return medoids
```

### Atualização da matriz de pertinência (Corrigir)

```python
def atualizacao_matriz_pertinencia(dados, medoids, m):
    matriz_distancias = np.linalg.norm(dados[:, np.newaxis] - medoids, axis=2)
    matriz_distancias = np.fmax(matriz_distancias, np.finfo(np.float64).eps)
    matriz_pertinencia_atualizada = np.zeros((dados.shape[0], medoids.shape[0]))

    for i in range(medoids.shape[0]):
        numerador = matriz_distancias[:, i] ** (1 / (m - 1))
        denominador = np.sum(numerador[:, np.newaxis] / matriz_distancias, axis=1)
        matriz_pertinencia_atualizada[:, i] = 1 / denominador

    return matriz_pertinencia_atualizada
```

### FCMdd (Corrigir)

```python
def fcmdd(dados, num_clusters, m=2, max_iter=1000, erro=1e-5):
    num_amostras = dados.shape[0]
    matriz_pertinencia = inicializacao_matriz_pertinencia(num_amostras, num_clusters)

    for _ in range(max_iter):
        medoids = atualizacao_medoids(dados, matriz_pertinencia, m)
        nova_matriz_pertinencia = atualizacao_matriz_pertinencia(dados, medoids, m)

        if np.linalg.norm(nova_matriz_pertinencia - matriz_pertinencia) < erro:
            break

        matriz_pertinencia = nova_matriz_pertinencia

    return medoids, matriz_pertinencia
```

### Índice de Rand Ajustado

```python
def indice_rand(labels, predicted_labels):
    return adjusted_rand_score(labels, predicted_labels)
```

### Experimento de Monte Carlo

```python
def experimento_monte_carlo(dados, labels, num_clusters, num_trials):
    indices_rand = []

    for trial in range(num_trials):
        medoids, matriz_pertinencia = fcmdd(dados, num_clusters)
        predicted_labels = np.argmax(matriz_pertinencia, axis=1)
        idx_rand = indice_rand(labels, predicted_labels)
        indices_rand.append(idx_rand)

    mean_rand_index = np.mean(indices_rand)
    std_rand_index = np.std(indices_rand)

    return mean_rand_index, std_rand_index
```

## Execução do método

```python
num_clusters = 3
num_trials = 100
media_indice_rand, dp_indice_rand = experimento_monte_carlo(dados, labels, num_clusters, num_trials)

print(f"Monte Carlo FCMdd Clustering Results ({num_trials} trials)")
print(f"Mean Rand Index: {media_indice_rand:.4f}") # Resultado esperado: 0.3589
print(f"Standard Deviation of Rand Index: {dp_indice_rand:.4f}")
```
